```python
# 체류시간 예측 실험
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BASE_PATH = os.getcwd()
DATA_PATH = BASE_PATH + r'\online_shoppers_intention.csv'

df = pd.read_csv(DATA_PATH, encoding='utf-8')
df.info()
df.head()
```

```python
def preprocess_data(df):
    df_preprocessed = df.copy()

    # 새로운 특성 생성
    df_preprocessed['Total_Duration'] = (
        df_preprocessed['Administrative_Duration'] + 
        df_preprocessed['Informational_Duration'] + 
        df_preprocessed['ProductRelated_Duration']
    )
    df_preprocessed['Total_Pages'] = (
        df_preprocessed['Administrative'] + 
        df_preprocessed['Informational'] + 
        df_preprocessed['ProductRelated']
    )
    df_preprocessed['Engagement_Rate'] = (
        df_preprocessed['PageValues'] / (
            df_preprocessed['BounceRates'] + 
            df_preprocessed['ExitRates'] + 1e-8
        )
    )

    # 범주형 변수 인코딩
    from sklearn.preprocessing import LabelEncoder

    le_month = LabelEncoder()
    le_visitor = LabelEncoder()
    
    df_preprocessed['Month_Encoded'] = le_month.fit_transform(df_preprocessed['Month'])
    df_preprocessed['VisitorType_Encoded'] = le_visitor.fit_transform(df_preprocessed['VisitorType'])
    df_preprocessed['Weekend_Encoded'] = df_preprocessed.fit_transform(df_preprocessed['Weekend']).astype(int)

    return df_preprocessed


# 데이터 전처리
df_preprocessed = preprocess_data(df)
print("전처리 완료")
print(f"새로운 특성 추가: Total_Duration, Total_Pages, Engagement_Rate")
print(f"범주형 변수 인코딩: Month, VisitorType, Weekend\n")
```
