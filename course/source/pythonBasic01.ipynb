## 파이썬 기본문법

### 자료형(문자열)

```python
num1 = 10
num2 = -20
print(num1, num2)
```

```python
num1 = 3.14
num2 = -2.718
print(num1, num2)
```

```python
str1 = 'Hello'
str2 = "Python"
print(str1, str2)
```

```python
# 멀티라인 문자열
str3 = '''This is
a multi-line
string'''
print(str3)
```

```python
text = "Python"
print(text[0])
print(text[-1])  # 뒤에서 첫 번째 문자
```

```python
text = "Python"
print(text[0:3])
print(text[::2])   # 2칸씩 건너뜀
```

```python
# upper(): 모든문자를 대문자로 변환, lower(): 모든 문자를 소문자로 변환
text = "Python"
print(text.upper())
print(text.lower())
```

```python
# find(): 특정 문자의 인덱스 반환(없으면 -1 반환), count(): 특정 문자의 개수 세기
text = "banana"
print(text.find('a'))
print(text.count('a'))
```

```python
# strip(): 문자열 양쪽 끝 공백 제거, lstrip() rstrip(): 왼쪽 또는 오른녹 공백만 제거거
text = " Hello World "
print(text)
print(text.strip())
```

```python
# split(): 특정 구분자로 문자열을 나누어 리스트로 반환
# join(): 리스트의 문자열을 특정 구분자로 연결하여 하나의 문자열로 반환환
text ="Python is fun"
words = text.split()
print(words)
joined_text =' '.join(words)
print(joined_text)
```

```python
# 예시: 파이썬 3.6이상, f-strings 방식
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

```python
# 구버전
text = "My name is %s and I am %d years old." %(name, age)
print(text)
```

```python
print("Hello\nPython")    # \n: 줄바꿈
print("탭을\t사용한\t예시")    # \t: 탭
print("백슬래시는 이렇게\\출력")   # \\: 백슬래시시
```

### 불린(bool)

```python
a = True
b = False
print(a and b)   # and: 두 조건이 모두 참일 때 참
print(a or b)    # or: 둘 중 하나라도 참이면 참
print(not a)    # not: 조건의 값을 반전시킴
```

```python
is_authenticated = True
has_permission = False
if is_authenticated and has_permission:
    print("접근 허용")
else:
    print("접근 거부")
```

### 리스트(list)

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])
```

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0:2])
```

```python
numbers = [3, 1, 4, 1, 5]
numbers.append(9)    # append: 요수추가 함수
print(numbers)
```

```python
numbers.sort()    # sort: 요소정렬 함수
print(numbers)
```

```python
numbers.remove(1)    #remove: 요소제거 함수
print(numbers)
```

```python
removed_item = numbers.pop(2)    # pop: 특정 인덱스 위치의 요소 추출 및 삭제
print(numbers)
print(removed_item)
```

```python
numbers.reverse()    # reverse: 리스트 뒤집기
print(numbers)
```

```python
count = numbers.count(5)    # count: 리스트 요소 개수 세기
print(count)
```

```python
new_numbers = numbers.copy()    # copy: 리스트 복사
print(new_numbers)
```

```python
numbers.extend([6, 7, 8])    # extend: 리스트 확장장
print(numbers)
```

```python
another_list = [10, 11, 12]
combined_list = numbers + another_list   # 각 리스트를 +로 연결
print(combined_list)
```

### 튜플(tuple)

```python
colors = ("red", "green", "blue")
print(colors[1])    # 인덱스로 접근
```

```python
colors = ("red", "green","blue")
print(colors[0:2])    # 마지막 숫자에서 -1
```

```python
point = (3, 4)
print("x 좌표:", point[0])
print("y 좌표:", point[1])
```

### if문

```python
score = 85
if score > 80:
    print("테스트에 합격하셨습니다.")
```

```python
number = 5
if number % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

```python
# if - elif - else 문
score = 85
if score >= 90:
    print("A학점")
elif score >= 80:
    print("B학점")
else:
    print("C학점")
```

```python
x, y = 10, 20
if x > 5 and y < 30:
    print("조건을 만족합니다.")
```

### for문

```python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num * 2)
```

```python
# range() 함수와 함께 사용 시
for i in range(1, 6):
    print(f"현재 값: {i}")
```

```python
for i in range(1, 4):
    for j in range(1, 3):
        print(f"i:{i}, j:{j}")
```

```python
items = ["A", "B", "C"]
for idx, item in enumerate(items):
    print(f"인덱스 {idx}:{item}")
```

### while문

```python
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1
```

```python
while True:
    user_input = input("종료하려면 'exit' 입력:")
    if user_input == "exit":
        break
```

```python
x = 5
while x > 0:
    print(x)
    x -= 1
else:
    print("반복 종료")
```

```python
# break & continue 사용
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue    # 짝수는 출력 건너뜀
    print(n)
    if n == 7:
        break    # 7에서 반복 종료
```

```python
password = "password123"
attempt = 0

while attempt < 3:
    user_password = input("비밀번호를 입력하세요: ")
    attempt += 1
    if user_password == password:
        print("비밀번호가 일치합니다.")
        break
    else:
        print("비밀번호가 틀렸습니다.")

if attempt == 3:
    print("비밀번호 입력 횟수를 초과했습니다.")
```
