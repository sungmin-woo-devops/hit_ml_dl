### while문


```python
count = 0
while count < 5:
  print(f"Count: {count}")
  count += 1
```

```python
# 무한루프 (ex. input조건)
while True:
    user_input = input("종료하려면 'exit' 입력: ")
    if user_input == "exit":
        break
```

```python
# 조건 검사 후 반복
number = 10
while number > 0:
    print(f"현재 값: {number}")
    number -= 2
```

```python
# 조건과 else 사용
x = 5
while x > 0:
    print(x)
    x -= 1
else:
    print("반복 종료")
```

```python
# 중첩 while문
outer = 1
while outer <= 3:
    inner = 1
    while inner <= 2:
        print(f"outer: {outer}, inner: {inner}")
        inner += 1
    outer += 1
```

```python
# break 와 continue
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue  # 짝수는 출력 건너뜀
    print(n)
    if n == 7:
        break  # 7에서 반복 종료
```

### 기본함수

```python
# 매개 변수가 없는 경우
def greet():
    print("안녕하세요!")
greet()
```

```python
# 매개 변수가 있는 경우
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

```python
# 기본값 매개변수 사용
def greet(name="홍길동님"):
    print(f"안녕하세요, {name}!")
greet()
greet("준호님")
```

```python
# 가변 매개변수 사용 (*args : 임의의 갯수의 숫자를 자유롭게 받아들임)
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result
print(multiply(2, 3, 4))
```

### 람다함수

```python
# 단순 계산
square = lambda x: x ** 2
print(square(5))
```

```python
# 여러 매개변수 사용
add = lambda a, b: a + b
print(add(3, 7))
```

```python
# 조건문 사용
check_even = lambda x: "짝수" if x % 2 == 0 else "홀수"
print(check_even(4))
```

```python
# 다른 함수를 인자로 사용
numbers = [5, 2, 9, 1]
sorted_numbers = sorted(numbers, key=lambda x: x)
print(sorted_numbers)
```

```python
# 참고 : sorted 함수 예시 (1)
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
sorted_list = sorted(my_list)
print(sorted_list)
```

```python
# 참고 : sorted 함수 예시 (2)
# 튜플 리스트를 두 번째 요소 기준으로 정렬
students = [('alice', 95), ('bob', 87), ('charlie', 92)]
sorted_students = sorted(students, key=lambda x: x[1])  # 점수 기준으로 정렬
print(sorted_students)
```

```python
# 참고 : sorted 함수 예시 (3)
# 문자열 리스트를 길이 기준으로 정렬
words = ['apple', 'banana', 'cherry', 'date']
sorted_words = sorted(words, key=len)  # 길이 기준으로 정렬
print(sorted_words)
```

```python
# 리스트 내 요소처리 : map(function, iterable) 함수는 필요할 때 계산되는 이터러블(반복순환) 객체에 사용
numbers = [1, 2, 3, 4]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

```python
# pop 인덱스를 통한 제거 예시
my_list = [1, 2, 3, 4, 5]

# 리스트의 마지막 요소를 빼내고 출력
last_item = my_list.pop()
print(last_item)

# 특정 인덱스의 요소를 빼내고 출력
third_item = my_list.pop(2)  # 인덱스는 0부터 시작
print(third_item)

print(my_list)
```

###문자열(String) 메서드

```python
# 문자열 메서드는 문자열을 직접 수정하지 않고 새 문자열 반환(기존 값은 불변성 유지)
text = "Hello"
print(text.lower())
print(text)
```

```python
# lower() 문자열을 소문자로 변환
text = "Hello World"
print(text.lower())
```

```python
# upper() 문자열을 대문자로 변환
text = "Hello World"
print(text.upper())
```

```python
# strip([chars]) 양쪽 끝의 공백 또는 특정문자 제거
text = "  Hello World  "
print(text.strip())
```

```python
# replace(old, new, [count]) 특정 문자열을 다른 문자열로 치환
text = "banana"
print(text.replace("a", "o"))
```

```python
# split([separator])  문자열을 구분자로 기준으로 분리하여 리스트로 반환
text = "apple,banana,cherry"
print(text.split(','))
```

```python
# join(iterable)  리스트나 튜플의 요소를 문자열로 결합
fruits = ['apple', 'banana', 'cherry']
print(', '.join(fruits))
```

```python
# find() 부분 문자열의 첫 번째 위치를 반환. 없으면 -1
text = "Hello World"
print(text.find("World"))
```

```python
# startswith(prefix) / endswith(suffix) 문자열이 특정 접두사/접미사로 시작하거나 끝나는지 확인
text = "Hello World"
print(text.startswith("Hello"))
print(text.endswith("World"))
```

```python
# isnumeric() 문자열이 숫자로만 구성되어 있는 지 확인
text = "12345"
print(text.isnumeric())
```

```python
# capitalize() 첫 번째 문자를 대문자로, 나머지는 소문자로 변환
text = "hello world"
print(text.capitalize())
```
